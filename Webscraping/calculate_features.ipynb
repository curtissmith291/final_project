```python
import pandas as pd
```

```python
# Read in descriptor words
key_words_df = pd.read_csv("beer_descriptors.csv")
```

```python
# Make column headers lowercase
key_words_df.columns = key_words_df.columns.str.lower()
```

```python
# Get list of key words
astringent = key_words_df.astringent.to_list()
body = key_words_df.body.to_list()
alcoholic = key_words_df.alcoholic.to_list()
bitter = key_words_df.bitter.to_list()
sweet = key_words_df.sweet.to_list()
sour = key_words_df.sour.to_list()
salty = key_words_df.salty.to_list()
fruity = key_words_df.fruity.to_list()
hoppy = key_words_df.hoppy.to_list()
spices = key_words_df.spices.to_list()
malty = key_words_df.malty.to_list()
```

```python
key_words_lod = {
    "astringent": key_words_df.astringent.to_list(),
    "body": key_words_df.body.to_list(),
    "alcoholic": key_words_df.alcoholic.to_list(),
    "bitter": key_words_df.bitter.to_list(),
    "sweet": key_words_df.sweet.to_list(),
    "sour": key_words_df.sour.to_list(),
    "salty": key_words_df.salty.to_list(),
    "fruity": key_words_df.fruity.to_list(),
    "hoppy": key_words_df.hoppy.to_list(),
    "spices": key_words_df.spices.to_list(),
    "malty": key_words_df.malty.to_list()
    }
```

```python
def get_astringent(reviews):
    astringent = 0
    reviews = reviews.replace("\n", "")
    split = reviews.split(" ")
    for word in split:
        if word in key_words_lod["astringent"]:
            astringent += 1
    return astringent

def get_body(reviews):
    body = 0
    reviews = reviews.replace("\n", "")
    split = reviews.split(" ")
    for word in split:
        if word in key_words_lod["body"]:
            body += 1
    return body

def get_alcoholic(reviews):
    alcoholic = 0
    reviews = reviews.replace("\n", "")
    split = reviews.split(" ")
    for word in split:
        if word in key_words_lod["alcoholic"]:
            alcoholic += 1
    return alcoholic

def get_bitter(reviews):
    bitter = 0
    reviews = reviews.replace("\n", "")
    split = reviews.split(" ")
    for word in split:
        if word in key_words_lod["bitter"]:
            bitter += 1
    return bitter

def get_sweet(reviews):
    sweet = 0
    reviews = reviews.replace("\n", "")
    split = reviews.split(" ")
    for word in split:
        if word in key_words_lod["sweet"]:
            sweet += 1
    return sweet

def get_sour(reviews):
    sour = 0
    reviews = reviews.replace("\n", "")
    split = reviews.split(" ")
    for word in split:
        if word in key_words_lod["sour"]:
            sour += 1
    return sour

def get_salty(reviews):
    salty = 0
    reviews = reviews.replace("\n", "")
    split = reviews.split(" ")
    for word in split:
        if word in key_words_lod["salty"]:
            salty += 1
    return salty

def get_fruity(reviews):
    fruity = 0
    reviews = reviews.replace("\n", "")
    split = reviews.split(" ")
    for word in split:
        if word in key_words_lod["fruity"]:
            fruity += 1
    return fruity

def get_hoppy(reviews):
    hoppy = 0
    reviews = reviews.replace("\n", "")
    split = reviews.split(" ")
    for word in split:
        if word in key_words_lod["hoppy"]:
            hoppy += 1
    return hoppy

def get_spices(reviews):
    spices = 0
    reviews = reviews.replace("\n", "")
    split = reviews.split(" ")
    for word in split:
        if word in key_words_lod["spices"]:
            spices += 1
    return spices

def get_malty(reviews):
    malty = 0
    reviews = reviews.replace("\n", "")
    split = reviews.split(" ")
    for word in split:
        if word in key_words_lod["malty"]:
            malty += 1
    return malty
```

```python
all_scraped_df["astringent"] = all_scraped_df.reviews.apply(get_astringent)
all_scraped_df["body"] = all_scraped_df.reviews.apply(get_body)
all_scraped_df["alcoholic"] = all_scraped_df.reviews.apply(get_alcoholic)
all_scraped_df["bitter"] = all_scraped_df.reviews.apply(get_bitter)
all_scraped_df["sweet"] = all_scraped_df.reviews.apply(get_sweet)
all_scraped_df["sour"] = all_scraped_df.reviews.apply(get_sour)
all_scraped_df["salty"] = all_scraped_df.reviews.apply(get_salty)
all_scraped_df["fruity"] = all_scraped_df.reviews.apply(get_fruity)
all_scraped_df["hoppy"] = all_scraped_df.reviews.apply(get_hoppy)
all_scraped_df["spices"] = all_scraped_df.reviews.apply(get_spices)
all_scraped_df["malty"] = all_scraped_df.reviews.apply(get_astringent)
```
